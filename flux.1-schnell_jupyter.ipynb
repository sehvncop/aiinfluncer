```python
# Install dependencies
!pip install -q diffusers transformers accelerate torch pillow

from diffusers import FluxImg2ImgPipeline
import torch
from PIL import Image

# Load pipeline
pipe = FluxImg2ImgPipeline.from_pretrained(
    "black-forest-labs/FLUX.1-schnell",
    torch_dtype=torch.float16
)
pipe.to("cuda")

# Load your reference image
ref_image = Image.open("/content/reference.jpg").resize((768, 1024))

# Your prompt
prompt = """
RAW photo, real person, natural beauty, 25 years old woman,
golden ratio facial proportions, natural skin texture,
visible pores, soft imperfections, warm approachable expression,
relatable girl next door, 85mm portrait lens,
soft natural window lighting, shallow depth of field,
not a model, real human energy, trustworthy face
"""

negative = """
plastic skin, perfect model, airbrushed, CGI, doll,
overprocessed, fake, artificial, Instagram filter,
supermodel, too perfect, symmetrical robot face
"""

# Generate — strength controls how much it changes your reference
# 0.3 = very close to original
# 0.6 = noticeable changes but same person energy
# 0.8 = mostly new but inspired by reference
result = pipe(
    prompt=prompt,
    image=ref_image,
    strength=0.55,        # sweet spot for twin effect
    num_inference_steps=4,
    guidance_scale=0.0,
).images[0]

result.save("/content/ai_twin_output.jpg")
print("Done! Download ai_twin_output.jpg from the sidebar")
```
